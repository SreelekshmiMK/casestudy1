```python
#importing the modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

1.Read the dataset to Python

```python
#reading the data from excel
data=pd.read_csv('cars_data.csv')
```

```python
#null values in the data
data.isnull()
```

3.Plor bar graph of female vs male buyers participated in the sales

```python
#counting total number of female and males participated in sale
data['Buyer Gender'].value_counts()
```

```python
#bar graph for female and male
plt.figure()
gender=['female','male']
number=[5052,4948]
cl=['green','red']
plt.bar(gender,number,color=cl)
plt.title('Bar Graph for female vs male')
plt.show()
```

```python
plt.figure()
plt.bar(data['Buyer Gender'],height=50,color=['blue','orange'])
plt.title('Female vs Male')
plt.xlabel('gender')
plt.ylabel('count')
```

4.Finding top 5 cars based on their sale price

```python
#grouping make and sale price
prd_sales=pd.DataFrame(data.groupby('Make').sum()['Sale Price'])
```

```python
prd_sales
```

```python
#sorting sale price from top list to bottom list
prd_sales=prd_sales.sort_values('Sale Price',ascending=False)
```

```python
prd_sales
```

```python
#top 5 cars based on their sale price
prd_sales[:5]
```

5.Finding least 5 cars based on their resell price

```python
#grouping make and resell price
prd_sales=pd.DataFrame(data.groupby('Make').sum()['Resell Price'])
```

```python
prd_sales
```

```python
#sorting for resell values from bottom to top
prd_sales=prd_sales.sort_values('Resell Price',ascending=True)
```

```python
prd_sales
```

```python
#least 5 cars based on resell price
prd_sales[:5]
```
